## **5.1 Loading Data**

Firstly, I will use monthly data for **univariate models**. Let's load daily data and convert it to monthly data. 

```python
# Load csv file
url='https://github.com/hangulde/Springboard/raw/main/Capstone%202/dataset_ready_for_EDA/all_features_eda.csv'
df = pd.read_csv(url, index_col='date')
df.index = pd.to_datetime(df.index) 
df_m=df.resample('MS').mean()
pm = df_m['pm25']
```

## **5.2 Is Monthly Data Stationary?**

```python
print('Results of Dickey-Fuller Test')

dftest = adfuller(pm)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lags Used', "Number of Observation Used"])

for key, value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
           
print(dfoutput)
```

Monthly data is not stationary. I will try diffent methods for making stationaty.

### **5.3 Differencing Method**

```python
print('Results of Dickey-Fuller Test - diff()-')

dftest = adfuller(pm.diff().dropna())
dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lags Used', "Number of Observation Used"])

for key, value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
           
print(dfoutput)

# Diff() doesn't work for making stationary.
```

```python
print('Results of Dickey-Fuller Test - diff(2)-')

dftest = adfuller(pm.diff(2).dropna())
dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lags Used', "Number of Observation Used"])

for key, value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
           
print(dfoutput)

# Diff(2) works for making stationary.
```

```python
# Plot of differenced data
pm.diff(2).dropna().plot()
```

I want to create two functions(**plot_forecast()**) and (**plot_forecast_cv()**) that shows the MAPE and plot of function. I create two different function because the latter for use the train-validation-test split.

```python
pred=None
def plot_forecast(test=y, pred = pred, name=None):
  print(f'MAPE for {name} model is {round(mean_absolute_percentage_error(y, pred), 3)}')
  plt.figure(figsize=(10,5))
  plt.plot(X.index, X, label='Train Data')
  plt.plot(y.index, y, label='Test Data')
  plt.plot(y.index, pred, label=f'{name} Model Forecast Data')
  plt.legend(loc='best')
  plt.xticks(rotation='vertical')
  plt.title(f'{name} Model for Monthly Data')
  plt.show()
```

```python
pred=None
forecast=None
def plot_forecast_cv(test=X_test, pred = pred, forecast=forecast, name=None):
  print(f'MAPE for {name} model is {round(mean_absolute_percentage_error(X_test, forecast), 3)}')
  plt.figure(figsize=(10,5))
  plt.plot(X_tr_val.index, X_tr_val, label='Train Data')
  plt.plot(X_test.index, X_test, label='Test Data')
  plt.plot(X_tr_val.index[10:], pred, label='Predicted Data')
  plt.plot(X_test.index, forecast, label= f'{name} Model Forecast Data')
  plt.legend(loc='best')
  plt.xticks(rotation='vertical')
  plt.title(f'{name} Model for Monthly Data')
  plt.show()
```

### **5.5.1 Naive Model**

This forecasting technique which assumes that the next expected point is equal to the last observed point is called ***Naive Method***.

```python
# create an array with the last value of train data set. 
pred = np.empty(len(y)) 
pred.fill(X[-1])
```

```python
plot_forecast(test=y, pred = pred, name='Naive')
```

### **5.5.2 Seasonal Naive**

Another method similar to naive model is useful for highly seasonal data. In this case, I set each forecast to be equal to the last observed value from the same season (e.g., the same month of the previous year).

```python
# take the last year's values.
pred = X.loc['2019-01-01':] 
# create a new predicton for two year
pred = pd.concat([pred, pred], axis=0)
```

```python
plot_forecast(test=y, pred = pred, name='Seasonal Naive')
```

### **5.5.3 Simple Average Model**

Such forecasting technique which forecasts the expected value equal to the average of all previously observed points is called Simple Average technique.

```python
#create an array with the average of train data set. 
pred = np.empty(len(y)) 
pred.fill(X.mean())
```

**Define Functions**

I will define two functions:
* **get_mape()**: Function for get the MAPE of ARIMA or SARIMA model
* **get_order()**: Learn the best params acoording to MAPE.

```python
# define a function for calculating MAPE
def get_mape(X, y, order=(0,0,0), seasonal_order=(0,0,0,0)):
          
    # define and fit the model
    model = SARIMAX(X, order=order, seasonal_order=seasonal_order, enforce_stationarity=False)
    model_fit = model.fit()
    
    # predict the future!
    pred = model_fit.forecast(len(y))
    
    # return the MAPE on our prediction
    return mean_absolute_percentage_error(y, pred)
```

```python
# Test our get_mape() function
print('MAPE: ', get_mape(X, y, (4,2,5)))

# same with previous model
```

```python
# define a function for getting order
def get_order (X, y):
    i = 0
    mapes = pd.DataFrame()  # holds model parameters and MAPEs
    # nested loop through all parameters.  
    for p in range(0,2):
        for d in [0,1,2]:  
            for q in range (0,3):
                for ps in range(0,2):
                    for ds in range(0,2):
                        for qs in range(0,2):
                            # build/fit/predict model and get the MAPE
                            mape = get_mape(X, y, order=(p,d,q), 
                                          seasonal_order=(ps,ds,qs,12))
                            
                            # save the parameters and MAPE
                            mapes = mapes.append(
                                pd.DataFrame({'p':p, 
                                              'd':d, 
                                              'q':q,
                                              'ps':ps,
                                              'ds':ds,
                                              'qs':qs,
                                              'mape':mape}, index=[i]))
                            i = i+1
             
    # sort our collection by MAPE and return the parameters with the lowest MAPE
    mapes = mapes.sort_values(by='mape', axis=0, ascending=True).reset_index()
    return mapes.iloc[0]
```

### **5.5.11 LSTM**

```python
# Load csv file
url='https://github.com/hangulde/Springboard/raw/main/Capstone%202/dataset_ready_for_EDA/all_features_eda.csv'
df = pd.read_csv(url, index_col='date')
df.index = pd.to_datetime(df.index) 
df_m=df.resample('MS').mean()
pm = df_m[['pm25']]
```

```python
pm
# In earlier notebooks, I checked whether the data is stationary or not. Daily PM2.5 is stationary but monthly is not.
```

```python
# Split train and test
train = pm.loc[:'2020-12-01']
test = pm.loc['2021-01-01':]
```

```python
train
```

LSTMs expect data to be within the scale of the activation function used by the network. The default activation function for LSTMs is the hyperbolic tangent (tanh), which outputs values between -1 and 1. This is the preferred range for the time series data.

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Fit the train data and transform both train and test
scaler.fit(train)
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
# Keras provides the TimeseriesGenerator that can be used to automatically 
# transform a univariate or multivariate time series dataset into a supervised learning problem.
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# Define Generator
n_input = 12
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
# number of samples
print('Samples: %d' % len(generator))
```

```python
# Check the generator
X, y = generator[2]
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')

# Now we can fit a model on this data and learn to map the input sequence to the output sequence.
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# fit model
model.fit(generator,epochs=70)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch = scaled_train[-12:]
```

```python
last_train_batch = last_train_batch.reshape((1, n_input, n_features))
```

```python
model.predict(last_train_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test)):
    
    # get the prediction value for the first batch
    current_pred = model.predict(current_batch)[0]
    
    # append the prediction into the array
    test_predictions.append(current_pred) 
    
    # use the prediction to update the batch and remove the first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
test_predictions
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```

```python
test['Predictions'] = true_predictions
```

```python
test.plot(figsize=(14,5))
```

```python
print(mean_absolute_percentage_error(test['pm25'], test['Predictions']))
```
